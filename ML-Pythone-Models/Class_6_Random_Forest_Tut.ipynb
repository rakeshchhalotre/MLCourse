```python
#labeling wine quality data as Good (quality>7) , Normal (quality 5-7) and Bad (quality<5)
# wine.loc[(wine.quantile<5),'taste'] = 'Bad' 
# wine.loc[(wine.quality>7),'taste'] = 'Good'
# wine.loc[(wine.quality>=5) & (wine.quality<=7)  ,'taste'] = 'Normal'

wine.loc[(wine['quality rate']<5),'taste'] = 'Bad' 
wine.loc[(wine['quality rate']>7),'taste'] = 'Good'
wine.loc[(wine['quality rate']>=5) & (wine['quality rate']<=7)  ,'taste'] = 'Normal'
```

```python
wine.head(15)
```

```python
#Initiating Random Forest Object
model_rf = RandomForestClassifier()
```

```python
X = wine.loc[:,:'alcohol']
y = wine.taste
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
model_rf.fit(X_train,y_train)
```

```python
y_predict = model_rf.predict(X_test)
```

```python
accuracy_score(y_test,y_predict)
```

```python
confusion_matrix(y_test,y_predict)
```

# Series

```python
# Series and DataFrame are data stuctures of Pandas
my_exp = [23,45,54,60,120, 50]
my_exp
```

```python
my_exp_series = pd.Series(my_exp)
my_exp_series
```

```python
#my_exp_series.index = ['day1','day2','day3','day4','day5','day6']

col = ['day1','day2','day3','day4','day5','day6']
my_exp_series.index = col
my_exp_series
```

# Retrival by Location

### we can retrieve value by 3 ways, as below

```python
# 1 # retrival by the label / index name
print(my_exp_series.day3)

# retrival by the index
print(my_exp_series[2])
```

```python
# 2 # Retrive  range by Slicing, you will get everyting
print(my_exp_series['day2':'day3'])

# Retrive  range by Slicing index, your starting point get excluded
print(my_exp_series[1:4])
```

```python
# 3 # Retrieve by list label

my_exp_series[['day1','day2','day3']]

# Retrieve by list index
my_exp_series[[5,2,4]]

```

###  Retrieve by Condition - SCALAR Filtering

```python
my_exp_series[my_exp_series >= 50]
```

```python
my_exp_series[(my_exp_series >= 40) & (my_exp_series <100) ]
```

```python
my_exp_series[(my_exp_series >= 40) & ((my_exp_series % 2) == 1) ]
```

# Data Treatment, Mugging , Preparation, Data Engineering
#### remove data problems ( missing data, duplicate, junk data...)

```python
data = pd.read_csv('mtcars_missing.csv')
data.head()
```

#### Missing value treatment

```python
data.info()
```

```python
#data.isnull()
data.isnull().sum()
```

```python
data.wt.str.contains('\?').sum()
```

# Replace using regular expression


```python
import numpy as np

data.iloc[:,1:].replace(r'[^0-9.]',np.nan, regex =True)
```

```python
data
```

```python
#data.wt.str.replace(' ?',np.nan)
```

```python
# To drop all the NaN value record
# If we have < 5% duplicate record we drop it
data.dropna()
data.head()
```

```python
data.shape
```

```python
(data.shape[0] -data.dropna().shape[0]) / data.shape[0]
```

```python
# below one drop missing(nan) value on UI and show not drop from data
# data = data.dropna()

# below will drop missing (nan) value from data
#data.dropna(inplace=true)
```

```python
#data.fillna(100)

# forward filling, fill record from above one ( before one)
data.fillna(method = 'ffill')
data.head()
```

```python

# bottom filling, fill missing value with the next value
data.fillna(method = 'bfill')
data.head()
```

```python
# mean filling, fill missing value with the mean of the column
data.fillna( data.mean())
data.head()
```

# Outliears - Extream values - remove those

```python
%matplotlib inline
data.plot(kind ='box')
```

```python
data[ data.mpg > (data.mpg.mean() + 3* data.mpg.std() )]
```

# Drop outliars from both side of the standatrd Deviation

```python
# Drop outliars from both side of the standatrd Deviation

data[ (data.mpg > (data.mpg.mean() - 3* data.mpg.std())) & (data.mpg < (data.mpg.mean() + 3* data.mpg.std() ))]
```

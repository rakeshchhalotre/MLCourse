### NUMPY  (Numerical Python) Tutorial

```python
import numpy as np
```

```python
a = np.array([5,8,6])
print("Array : ",a)
print("N-Dimension : ",a.ndim)
print("Array Shape  : ",a.shape)
print("Size : ",a.size)
print("Element Data Type : ",a.dtype)
print("Data Type : ",type(a))
```

```python
a = np.array([
              [5,8,6],
              [12,16,15]
            ])

print("Array : \n",a)
print("N-Dimension : ",a.ndim)
print("Array Shape  : ",a.shape)
print("Size : ",a.size)
print("Element Data Type : ",a.dtype)
print("Data Type : ",type(a))
```

### Array Properties
- Array are homogeneous
- can't have inconsistent elements number

```python
a = np.array([[
              [5,8,6],
              [12,16,15],
              [5,8,6],
              [5,8,6]
            ],
            [
              [50,84,26],
              [25,36,45],
              [5,8,6],
              [5,8,6]
            ]])

print("Array : \n",a)
print("N-Dimension : ",a.ndim)
print("Array Shape  : ",a.shape)
print("Size : ",a.size)
print("Element Data Type : ",a.dtype)
print("Data Type : ",type(a))
```

```python
a[1][0][1]
```

###  np.arange() function, Array Range

```python
a = np.arange(10)
print("Array : \n",a)
print("N-Dimension : ",a.ndim)
print("Array Shape : ",a.shape)
print("Size : ",a.size)
print("Element Data Type : ",a.dtype)
print("Data Type : ",type(a))
```

```python
a = np.arange(10,20)
print("Array : \n",a)
print("N-Dimension : ",a.ndim)
print("Array Shape : ",a.shape)
print("Size : ",a.size)
print("Element Data Type : ",a.dtype)
print("Data Type : ",type(a))
```

```python
# arange(start,stop,step)
a = np.arange(10,50,5)
print("Array : \n",a)
print("N-Dimension : ",a.ndim)
print("Array Shape : ",a.shape)
print("Size : ",a.size)
```

### np.linspace(start,stop,number of elements)

```python
# linspace(start,stop,number of elements)
a = np.linspace(10,50,5)
print("Array : \n",a)
print("N-Dimension : ",a.ndim)
print("Array Shape : ",a.shape)
print("Size : ",a.size)
```

### np.reshape(x,y,z,..) 

```python
# linspace(start,stop,number of elements)
a = np.arange(20).reshape(2,2,5)
print("Array : \n",a)
print("N-Dimension : ",a.ndim)
print("Array Shape : ",a.shape)
print("Size : ",a.size)
```

```python
# Pandas => Panel data 
import numpy as np
import pandas as pd

from pandas import Series , DataFrame
```

```python
# !pwd -> give the present working directory path
# !lf -> list all files under directory

cars = pd.read_csv('pima-data-csv.csv')
cars.head()
cars
```

```python
col = ['carmodel', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'vs', 'am', 'gear']
cars.columns=col
cars
```

```python
# conditiond using Headers
cars.carmodel
```

```python
cars[cars.carmodel>10]
```

```python
# multiple filter conditions
cars[(cars.carmodel>10) & (cars.disp < 30)]
```

```python
#filter missing values using numpy nan

cars.disp[cars.disp <1 ] = np.nan
cars
```

```python
# fill forward method
# take before value and fill into the zero or notAvailable value
cars.fillna(method = 'ffill')
```

```python
# fill backword method
# take after value and fill into the zero or notAvailable value
cars.fillna(method = 'bfill')
```

```python
# fill Mean for all NaN, which we filled using np.nan
cars.fillna(cars.mean())
```

```python
cars
```

```python
# drop the rercords having NaN (missing value record)
cars.dropna()
```
